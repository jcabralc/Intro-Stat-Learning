###  9. (e) 

```python
plt.figure(figsize=(15,5))

plt.subplot(221)
ax1 = plt.scatter(auto.mpg, auto.cylinders, alpha=0.5)
plt.xlabel('mpg')
plt.ylabel('cylinders')
plt.title('mpg to cylinders')

plt.subplot(222)
ax2 = plt.scatter(auto.mpg, auto.horsepower, alpha=0.5)
plt.xlabel('mpg')
plt.ylabel('horsepower')
plt.title('mpg to horsepower')

plt.subplot(223)
ax3 = plt.scatter(auto.mpg, auto.acceleration, alpha=0.5)
plt.xlabel('mpg')
plt.ylabel('acceleration')
plt.title('mpg to acceleration')

plt.subplot(224)
ax3 = plt.scatter(auto.mpg, auto.year, alpha=0.5)
plt.xlabel('mpg')
plt.ylabel('year')
plt.title('mpg per year')

plt.tight_layout()

plt.show()
```

more cylinders cosume less mpg

more horsepower, less mpg

less acceleration, less mpg

cars become more efficient over time
